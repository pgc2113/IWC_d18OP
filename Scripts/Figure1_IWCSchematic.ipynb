# Load Packages

```python
# Basic Plotting
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import matplotlib.patches as mpatches
import seaborn as sns

# Creating Maps
import cartopy.crs as ccrs
import cartopy.feature as cfeature

# Image Resolution on MacOS
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
```

# IWC Location Maps

```python
# Plotting Maps
fig = plt.figure(figsize=(24,14))
gs = gridspec.GridSpec(nrows=1, ncols=3)

# Tick labels
x_tick_labels = [u'40\N{DEGREE SIGN}E', u'60\N{DEGREE SIGN}E',u'80\N{DEGREE SIGN}E',
                u'100\N{DEGREE SIGN}E', u'120\N{DEGREE SIGN}E']
y_tick_labels = [u'30\N{DEGREE SIGN}S', u'15\N{DEGREE SIGN}S',u'0\N{DEGREE SIGN}',
                 u'15\N{DEGREE SIGN}N', u'30\N{DEGREE SIGN}N']

extent = [25, 120, -40, 30]
yticks = [-30,-15,0,15,30]
xticks = [40,60,80,100,120]

# Negative IWC 
ax0 = fig.add_subplot(gs[0], projection=ccrs.PlateCarree(central_longitude=0))
ax0.set_extent(extent, ccrs.PlateCarree())
ax0.set_yticks(yticks)
ax0.set_yticklabels(y_tick_labels, fontsize = 12, fontweight='bold')
ax0.set_xticks(xticks)
ax0.set_xticklabels(x_tick_labels, fontsize = 12, fontweight='bold', rotation = 45)
ax0.add_feature(cfeature.COASTLINE.with_scale('50m'))
ax0.add_feature(cfeature.BORDERS)
ax0.set_title('Positive Phase', fontsize = 28, fontweight = 'bold', color = 'darkblue')
ax0.set_xlabel('', fontsize = 8)
ax0.set_ylabel('', fontsize = 8)

ax0.add_patch(mpatches.Rectangle(xy = [60, -5], width = 30, height = 10, facecolor = 'none', edgecolor = 'darkgreen', transform = ccrs.PlateCarree())) # Ueq Location
ax0.add_patch(mpatches.Rectangle(xy = [90, -7.5], width = 20, height = 15, facecolor = 'none', edgecolor = 'dodgerblue', # EIO (Ascending Arm)
                                 fill = True, fc = 'dodgerblue', alpha = 0.4, 
                                 transform=ccrs.PlateCarree()))
ax0.add_patch(mpatches.Rectangle(xy = [35, -7.5], width = 10, height = 15, facecolor = 'none', edgecolor = 'tomato', # WIO (Descending Arm)
                                 fill = True, fc = 'tomato', alpha = 0.4, 
                                 transform = ccrs.PlateCarree()))
# Neutral IWC
ax1 = fig.add_subplot(gs[1], projection=ccrs.PlateCarree(central_longitude=0))
ax1.set_extent(extent, ccrs.PlateCarree())
ax1.set_yticks(yticks)
ax1.set_yticklabels(y_tick_labels, fontsize = 12, fontweight='bold')
ax1.set_xticks(xticks)
ax1.set_xticklabels(x_tick_labels, fontsize = 12, fontweight='bold', rotation = 45)
ax1.add_feature(cfeature.COASTLINE.with_scale('50m'))
ax1.add_feature(cfeature.BORDERS)
ax1.set_title('Neutral Phase', fontsize = 28, fontweight = 'bold', color = 'black')
ax1.set_xlabel('', fontsize = 12, fontweight='bold')
ax1.set_ylabel('', fontsize = 12, fontweight='bold')

ax1.add_patch(mpatches.Rectangle(xy = [60, -5], width = 30, height = 10, facecolor = 'none', edgecolor = 'darkgreen', transform = ccrs.PlateCarree())) # Ueq Location
ax1.add_patch(mpatches.Rectangle(xy = [90, -7.5], width = 20, height = 15, facecolor = 'none', edgecolor = 'dodgerblue', # EIO (Ascending Arm)
                                 fill = True, fc = 'dodgerblue', alpha = 0.4, 
                                 transform=ccrs.PlateCarree()))
ax1.add_patch(mpatches.Rectangle(xy = [35, -7.5], width = 10, height = 15, facecolor = 'none', edgecolor = 'tomato', # WIO (Descending Arm)
                                 fill = True, fc = 'tomato', alpha = 0.4, 
                                 transform = ccrs.PlateCarree()))

# Positive IWC 
ax2 = fig.add_subplot(gs[2], projection=ccrs.PlateCarree(central_longitude=0))
ax2.set_extent(extent, ccrs.PlateCarree())
ax2.set_yticks(yticks)
ax2.set_yticklabels(y_tick_labels, fontsize = 12, fontweight='bold')
ax2.set_xticks(xticks)
ax2.set_xticklabels(x_tick_labels, fontsize = 12, fontweight='bold', rotation = 45)
ax2.add_feature(cfeature.COASTLINE.with_scale('50m'))
ax2.add_feature(cfeature.BORDERS)
ax2.set_title('Negative Phase', fontsize = 28, fontweight = 'bold', color = 'darkred')
ax2.set_xlabel('', fontsize = 12, fontweight='bold')
ax2.set_ylabel('', fontsize = 12, fontweight='bold')

ax2.add_patch(mpatches.Rectangle(xy = [60, -5], width = 30, height = 10, facecolor = 'none', edgecolor = 'darkgreen', transform = ccrs.PlateCarree())) # Ueq Location
ax2.add_patch(mpatches.Rectangle(xy = [90, -7.5], width = 20, height = 15, facecolor = 'none', edgecolor = 'tomato', # EIO (Descending Arm)
                                 fill = True, fc = 'tomato', alpha = 0.4, 
                                 transform=ccrs.PlateCarree()))
ax2.add_patch(mpatches.Rectangle(xy = [35, -7.5], width = 10, height = 15, facecolor = 'none', edgecolor = 'dodgerblue', # WIO (Ascending Arm)
                                 fill = True, fc = 'dodgerblue', alpha = 0.4, 
                                 transform = ccrs.PlateCarree()))

plt.tight_layout()
plt.savefig('IndianOceanBasinImage.pdf',bbox_inches='tight',dpi=300)
```
